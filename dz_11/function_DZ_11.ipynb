# Даны две функции:
f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
Требуется:
1. Найти координаты точек пересечения
2. Построить графики функций в одной системе координат
3. Построить график функции пересечения

# График функции f(x) = x^3 - 50x

```python
from sympy.plotting import plot
from sympy import *
x = Symbol('x')
fx = x**3 - 50*x
plot(fx, line_color='red')
```

# График функции g(x) = -x^4 + 88x^2 - 241

```python
x = Symbol('x')
gx = -x**4 + 88*x**2 - 241
plot(gx, line_color='blue')
```

# Корни

```python
fxg = fx - gx
recal = solve(fxg)
recal[1]
```

# 1. Координаты точек пересечения

```python
coord = {}
c = 1
for z in recal:
    coord[c] = {'x': round(N(z), 5), 'y': round(N(-1*z**4 + 88*z**2 - 241), 5)}
    c += 1
coord
```

# 2. Графики функций в одной системе координат

```python
plot(fx, gx)
```

# 3. График функции пересечения

```python
plot(fx, gx, fxg)
```

```python
plot(fxg, line_color='green')
```
